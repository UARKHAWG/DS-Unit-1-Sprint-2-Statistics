```
A variable that has a normal distribution is one where the mean = median = mode. A normal distribution is theoretical, which means that in nature, nothing will strictly follow this distribution.But many variables approximate it, such as the heights of adult humans, human baby birth weights, and the heights of the same variety of pine trees, in addition to many other things.

The normal distribution is known as a bell curve because it has the shape of a bell. Using the numpy normal distribution, we can draw a sample population with a specified mean (center point of the "bell") and standard deviation (width or "spread" of the curve) and plot.
```
2. What does the Central Limit Theorem tell us?
```
The central limit theorem states that if you have a population with mean μ and standard deviation σ and take sufficiently large random samples from the population with replacement , then the distribution of the sample means will be approximately normally distributed.
**The symbol 'μ' represents the population mean**
```
3. What does the Law of Large Numbers tell us?
```
The law of large numbers states that an observed sample average from a large sample will be close to the true population average and that it will get closer the larger the sample.
```
4. What is the t-distribution and how does it relate to the Normal distribution?
```
A t-test is based on a t-value. When you perform a t-test for a single study, you obtain a single t-value. If you drew lots of random samples of the same size from the same population, performed a t-test each time (to obtain a t-value), you could then plot a distribution of all the t-values. This type of distribution is called a sampling distribution and, in this case, a t-distribution.

The T distribution is similar to the normal distribution, just with fatter tails. Both assume a normally distributed population. T distributions have higher kurtosis than normal distributions. The probability of getting values very far from the mean is larger with a T distribution than a normal distribution.
```

8. What is a p-value?
```
The **p-value** is the probability that, **GIVEN THE NULL HYPOTHESIS IS TRUE**, we would observe the current set of data. This "chance" is called the p-value - the probability of observing the test-statistic, as is, given the Null hypothesis is true (e.g., this grouping of numbers is the result of random chance and not significant). And this probability is calculated under the assumption of a certain probability distribution (that the test statistic is generated from).
```
9. How do you use the p-value and alpha level to determine if you should reject or fail to reject the null hypothesis?
```
If your p-value is less than a chosen significance level $\alpha$, then you may choose to reject the null hypothesis and accept there is reasonable evidence to support the alternative hypothesis.
If your p-value is more than a chosen significance level $\alpha$, then you may choose to fail to reject the null hypothesis.
```

15. Explain the difference betwen a t-test and a chi-square test.
```
A t-test and a chi-square test are two different statistical tests used in hypothesis testing, and they serve different purposes and are applied to different types of data. Here are the key differences between these two tests:

Purpose:

T-test: A t-test is used to compare the means of two groups or to determine if a sample mean is significantly different from a known population mean. It is primarily used for continuous numerical data, such as height, weight, test scores, etc.
Chi-square test: A chi-square test, on the other hand, is used to assess the association or independence between categorical variables. It helps determine if there is a significant relationship between two categorical variables or if the observed frequency distribution differs from the expected distribution.
Type of data:

T-test: It is used with continuous data, which is measured on an interval or ratio scale. The data can be unpaired (independent samples t-test) or paired (paired samples t-test).
Chi-square test: It is used with categorical data, which can be nominal or ordinal. It deals with counts or frequencies in categories and doesn't consider the magnitude or order of the categories.
Test statistics:

T-test: The test statistic for a t-test is calculated based on the difference in means and their standard errors. The t-statistic measures how many standard errors the sample means are away from each other or the population mean.
Chi-square test: The test statistic for a chi-square test is calculated based on the differences between observed and expected frequencies in a contingency table. It assesses the degree of association between categorical variables.
Types of t-tests:

T-test has various forms, including:
Independent samples t-test: Compares the means of two independent groups.
Paired samples t-test: Compares the means of two related groups.
One-sample t-test: Compares a sample mean to a known population mean.
Chi-square test has different types, including:
Chi-square test for independence: Examines the association between two categorical variables in a contingency table.
Chi-square goodness-of-fit test: Tests if observed frequencies match expected frequencies in a single categorical variable.
Chi-square test for homogeneity: Determines if the distributions of a categorical variable are the same across multiple groups.
In summary, t-tests are used for comparing means in continuous data, while chi-square tests are used for examining associations between categorical variables. The choice of which test to use depends on the nature of your data and the research question you want to answer.
```
